```python
import math
import warnings
from pathlib import Path
import cv2
import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn.functional as F
import seaborn as sns
import torchvision
from ultralytics import YOLO
# Load a model
model = None
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')

def _get_covariance_matrix(boxes):
    """
    Generating covariance matrix from obbs.

    Args:
        boxes (torch.Tensor): A tensor of shape (N, 5) representing rotated bounding boxes, with xywhr format.

    Returns:
        (torch.Tensor): Covariance metrixs corresponding to original rotated bounding boxes.
    """
    # Gaussian bounding boxes, ignore the center points (the first two columns) because they are not needed here.
    gbbs = torch.cat((boxes[:, 2:4].pow(2) / 12, boxes[:, 4:]), dim=-1)
    a, b, c = gbbs.split(1, dim=-1)
    cos = c.cos()
    sin = c.sin()
    cos2 = cos.pow(2)
    sin2 = sin.pow(2)
    return a * cos2 + b * sin2, a * sin2 + b * cos2, (a - b) * cos * sin
def txt2point(label_txt,image_shape):
    (h,w) = image_shape
    points = []
    with open(label_txt, 'r') as file:
        datas = file.readlines()
    for data in datas:
        cordi = [float(cor) for cor in data.strip().split(' ')[1:]]
        x = [round(x*w) for x in cordi[::2]]
        y = [round(y*h) for y in cordi[1::2]]
        point = np.array([[x,y] for x,y in zip(x,y)],dtype=np.int32)
        # mask = cv2.fillPoly(mask, pts=[points], color=255)
        # masks.append(mask)
        points.append(point)
    return points

def probiou(obb1, obb2, CIoU=False, eps=1e-7):
    """
    Calculate the prob IoU between oriented bounding boxes, https://arxiv.org/pdf/2106.06072v1.pdf.

    Args:
        obb1 (torch.Tensor): A tensor of shape (N, 5) representing ground truth obbs, with xywhr format.
        obb2 (torch.Tensor): A tensor of shape (N, 5) representing predicted obbs, with xywhr format.
        eps (float, optional): A small value to avoid division by zero. Defaults to 1e-7.

    Returns:
        (torch.Tensor): A tensor of shape (N, ) representing obb similarities.
    """
    x1, y1 = obb1[..., :2].split(1, dim=-1)
    x2, y2 = obb2[..., :2].split(1, dim=-1)
    a1, b1, c1 = _get_covariance_matrix(obb1)
    a2, b2, c2 = _get_covariance_matrix(obb2)

    t1 = (
        ((a1 + a2) * (y1 - y2).pow(2) + (b1 + b2) * (x1 - x2).pow(2)) / ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2) + eps)
    ) * 0.25
    t2 = (((c1 + c2) * (x2 - x1) * (y1 - y2)) / ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2) + eps)) * 0.5
    t3 = (
        ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2))
        / (4 * ((a1 * b1 - c1.pow(2)).clamp_(0) * (a2 * b2 - c2.pow(2)).clamp_(0)).sqrt() + eps)
        + eps
    ).log() * 0.5
    bd = (t1 + t2 + t3).clamp(eps, 100.0)
    hd = (1.0 - (-bd).exp() + eps).sqrt()
    iou = 1 - hd
    if CIoU:  # only include the wh aspect ratio part
        w1, h1 = obb1[..., 2:4].split(1, dim=-1)
        w2, h2 = obb2[..., 2:4].split(1, dim=-1)
        v = (4 / math.pi**2) * ((w2 / h2).atan() - (w1 / h1).atan()).pow(2)
        with torch.no_grad():
            alpha = v / (v - iou + (1 + eps))
        return iou - v * alpha  # CIoU
    return iou
def batch_probiou(obb1, obb2, eps=1e-7):
    """
    Calculate the prob IoU between oriented bounding boxes, https://arxiv.org/pdf/2106.06072v1.pdf.

    Args:
        obb1 (torch.Tensor | np.ndarray): A tensor of shape (N, 5) representing ground truth obbs, with xywhr format.
        obb2 (torch.Tensor | np.ndarray): A tensor of shape (M, 5) representing predicted obbs, with xywhr format.
        eps (float, optional): A small value to avoid division by zero. Defaults to 1e-7.

    Returns:
        (torch.Tensor): A tensor of shape (N, M) representing obb similarities.
    """
    obb1 = torch.from_numpy(obb1) if isinstance(obb1, np.ndarray) else obb1
    obb2 = torch.from_numpy(obb2) if isinstance(obb2, np.ndarray) else obb2

    x1, y1 = obb1[..., :2].split(1, dim=-1)
    x2, y2 = (x.squeeze(-1)[None] for x in obb2[..., :2].split(1, dim=-1))
    a1, b1, c1 = _get_covariance_matrix(obb1)
    a2, b2, c2 = (x.squeeze(-1)[None] for x in _get_covariance_matrix(obb2))

    t1 = (
        ((a1 + a2) * (y1 - y2).pow(2) + (b1 + b2) * (x1 - x2).pow(2)) / ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2) + eps)
    ) * 0.25
    t2 = (((c1 + c2) * (x2 - x1) * (y1 - y2)) / ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2) + eps)) * 0.5
    t3 = (
        ((a1 + a2) * (b1 + b2) - (c1 + c2).pow(2))
        / (4 * ((a1 * b1 - c1.pow(2)).clamp_(0) * (a2 * b2 - c2.pow(2)).clamp_(0)).sqrt() + eps)
        + eps
    ).log() * 0.5
    bd = (t1 + t2 + t3).clamp(eps, 100.0)
    hd = (1.0 - (-bd).exp() + eps).sqrt()
    return 1 - hd


def smooth_BCE(eps=0.1):
    """
    Computes smoothed positive and negative Binary Cross-Entropy targets.

    This function calculates positive and negative label smoothing BCE targets based on a given epsilon value.
    For implementation details, refer to https://github.com/ultralytics/yolov3/issues/238#issuecomment-598028441.

    Args:
        eps (float, optional): The epsilon value for label smoothing. Defaults to 0.1.

    Returns:
        (tuple): A tuple containing the positive and negative label smoothing BCE targets.
    """
    return 1.0 - 0.5 * eps, 0.5 * eps
def bbox_iou(box1, box2, xywh=True, GIoU=False, DIoU=False, CIoU=False, eps=1e-7):
    """
    Calculate Intersection over Union (IoU) of box1(1, 4) to box2(n, 4).

    Args:
        box1 (torch.Tensor): A tensor representing a single bounding box with shape (1, 4).
        box2 (torch.Tensor): A tensor representing n bounding boxes with shape (n, 4).
        xywh (bool, optional): If True, input boxes are in (x, y, w, h) format. If False, input boxes are in
                               (x1, y1, x2, y2) format. Defaults to True.
        GIoU (bool, optional): If True, calculate Generalized IoU. Defaults to False.
        DIoU (bool, optional): If True, calculate Distance IoU. Defaults to False.
        CIoU (bool, optional): If True, calculate Complete IoU. Defaults to False.
        eps (float, optional): A small value to avoid division by zero. Defaults to 1e-7.

    Returns:
        (torch.Tensor): IoU, GIoU, DIoU, or CIoU values depending on the specified flags.
    """

    # Get the coordinates of bounding boxes
    if xywh:  # transform from xywh to xyxy
        (x1, y1, w1, h1), (x2, y2, w2, h2) = box1.chunk(4, -1), box2.chunk(4, -1)
        w1_, h1_, w2_, h2_ = w1 / 2, h1 / 2, w2 / 2, h2 / 2
        b1_x1, b1_x2, b1_y1, b1_y2 = x1 - w1_, x1 + w1_, y1 - h1_, y1 + h1_
        b2_x1, b2_x2, b2_y1, b2_y2 = x2 - w2_, x2 + w2_, y2 - h2_, y2 + h2_
    else:  # x1, y1, x2, y2 = box1
        b1_x1, b1_y1, b1_x2, b1_y2 = box1.chunk(4, -1)
        b2_x1, b2_y1, b2_x2, b2_y2 = box2.chunk(4, -1)
        w1, h1 = b1_x2 - b1_x1, b1_y2 - b1_y1 + eps
        w2, h2 = b2_x2 - b2_x1, b2_y2 - b2_y1 + eps

    # Intersection area
    inter = (b1_x2.minimum(b2_x2) - b1_x1.maximum(b2_x1)).clamp_(0) * (
        b1_y2.minimum(b2_y2) - b1_y1.maximum(b2_y1)
    ).clamp_(0)

    # Union Area
    union = w1 * h1 + w2 * h2 - inter + eps

    # IoU
    iou = inter / union
    if CIoU or DIoU or GIoU:
        cw = b1_x2.maximum(b2_x2) - b1_x1.minimum(b2_x1)  # convex (smallest enclosing box) width
        ch = b1_y2.maximum(b2_y2) - b1_y1.minimum(b2_y1)  # convex height
        if CIoU or DIoU:  # Distance or Complete IoU https://arxiv.org/abs/1911.08287v1
            c2 = cw.pow(2) + ch.pow(2) + eps  # convex diagonal squared
            rho2 = (
                (b2_x1 + b2_x2 - b1_x1 - b1_x2).pow(2) + (b2_y1 + b2_y2 - b1_y1 - b1_y2).pow(2)
            ) / 4  # center dist**2
            if CIoU:  # https://github.com/Zzh-tju/DIoU-SSD-pytorch/blob/master/utils/box/box_utils.py#L47
                v = (4 / math.pi**2) * ((w2 / h2).atan() - (w1 / h1).atan()).pow(2)
                with torch.no_grad():
                    alpha = v / (v - iou + (1 + eps))
                return iou - (rho2 / c2 + v * alpha)  # CIoU
            return iou - rho2 / c2  # DIoU
        c_area = cw * ch + eps  # convex area
        return iou - (c_area - union) / c_area  # GIoU https://arxiv.org/pdf/1902.09630.pdf
    return iou  # IoU

def box_iou(box1, box2, eps=1e-7):
    """
    Calculate intersection-over-union (IoU) of boxes. Both sets of boxes are expected to be in (x1, y1, x2, y2) format.
    Based on https://github.com/pytorch/vision/blob/master/torchvision/ops/boxes.py

    Args:
        box1 (torch.Tensor): A tensor of shape (N, 4) representing N bounding boxes.
        box2 (torch.Tensor): A tensor of shape (M, 4) representing M bounding boxes.
        eps (float, optional): A small value to avoid division by zero. Defaults to 1e-7.

    Returns:
        (torch.Tensor): An NxM tensor containing the pairwise IoU values for every element in box1 and box2.
    """

    # NOTE: Need .float() to get accurate iou values
    # inter(N,M) = (rb(N,M,2) - lt(N,M,2)).clamp(0).prod(2)
    (a1, a2), (b1, b2) = box1.float().unsqueeze(1).chunk(2, 2), box2.float().unsqueeze(0).chunk(2, 2)
    inter = (torch.min(a2, b2) - torch.max(a1, b1)).clamp_(0).prod(2)

    # IoU = inter / (area1 + area2 - inter)
    return inter / ((a2 - a1).prod(2) + (b2 - b1).prod(2) - inter + eps)
def iou_smask(mask1, mask2, eps=1e-7):
    mask2_expanded = mask2.sum(1)[None]
    intersection = torch.matmul(mask1, mask2.T).clamp_(0)
    res = intersection / (mask2_expanded + 1e-7)
    return res
def mask_iou(mask1, mask2, eps=1e-7):
    """
    Calculate masks IoU.

    Args:
        mask1 (torch.Tensor): A tensor of shape (N, n) where N is the number of ground truth objects and n is the
                        product of image width and height.
        mask2 (torch.Tensor): A tensor of shape (M, n) where M is the number of predicted objects and n is the
                        product of image width and height.
        eps (float, optional): A small value to avoid division by zero. Defaults to 1e-7.

    Returns:
        (torch.Tensor): A tensor of shape (N, M) representing masks IoU.
    """
    intersection = torch.matmul(mask1, mask2.T).clamp_(0)
    union = (mask1.sum(1)[:, None] + mask2.sum(1)[None]) - intersection  # (area1 + area2) - intersection
    return intersection / (union + eps)
class ConfusionMatrix:
    """
    A class for calculating and updating a confusion matrix for object detection and classification tasks.

    Attributes:
        task (str): The type of task, either 'detect' or 'classify'.
        matrix (np.ndarray): The confusion matrix, with dimensions depending on the task.
        nc (int): The number of classes.
        conf (float): The confidence threshold for detections.
        iou_thres (float): The Intersection over Union threshold.
    """

    def __init__(self, nc, conf=0.25, iou_thres=0.45,mask_accept=0.8, task="detect"):
        """Initialize attributes for the YOLO model."""
        self.task = task
        self.matrix = np.zeros((nc + 1, nc + 1)) if self.task == "detect" else np.zeros((nc, nc))
        self.matrix_mask = np.zeros((nc + 1, nc + 1)) if self.task == "detect" else np.zeros((nc, nc))
        self.nc = nc  # number of classes
        self.mask_accept = mask_accept
        self.conf = 0.25 if conf in {None, 0.001} else conf  # apply 0.25 if default val conf is passed
        self.iou_thres = iou_thres
        self.sum_correct = 0
        self.sum_shortage_iou = 0
        self.sum_shortage = 0
        self.sum_surplus = 0
        self.total_object = 0
        self.total_mask_predict = 0
        self.total_mask_accept = 0

        self.name_surplus = []
        self.name_shortage = []
    def process_cls_preds(self, preds, targets):
        """
        Update confusion matrix for classification task.

        Args:
            preds (Array[N, min(nc,5)]): Predicted class labels.
            targets (Array[N, 1]): Ground truth class labels.
        """
        preds, targets = torch.cat(preds)[:, 0], torch.cat(targets)
        for p, t in zip(preds.cpu().numpy(), targets.cpu().numpy()):
            self.matrix[p][t] += 1
    def process_batch_mask(self, mask_detection, gt_cls,mask_gt):
        """
        Update confusion matrix for object detection task.

        Args:
            detections (Array[N, 6] | Array[N, 7]): Detected bounding boxes and their associated information.
                                        Each row should contain (x1, y1, x2, y2, conf, class)
                                        or with an additional element `angle` when it's obb.
            gt_bboxes (Array[M, 4]| Array[N, 5]): Ground truth bounding boxes with xyxy/xyxyr format.
            gt_cls (Array[M]): The class labels.
        """
        if gt_cls.shape[0] == 0:  # Check if labels is empty
            if mask_detection is not None:
                mask_detection = mask_detection[mask_detection[:, -2] > self.conf]
                detection_classes = mask_detection[:, -1].int()
                for dc in detection_classes:
                    self.matrix_mask[dc, self.nc] += 1  # false positives
            return
        if mask_detection is None:
            gt_classes = gt_cls.int()
            for gc in gt_classes:
                self.matrix_mask[self.nc, gc] += 1  # background FN
            return

        mask_detection = mask_detection[mask_detection[:, -2] > self.conf]
        gt_classes = gt_cls.int()
        detection_classes = mask_detection[:, -1].int()

        iou = box_iou(mask_gt, mask_detection[:, :-2])
        x = torch.where(iou > self.iou_thres)
        if x[0].shape[0]:
            matches = torch.cat((torch.stack(x, 1), iou[x[0], x[1]][:, None]), 1).cpu().numpy()
            if x[0].shape[0] > 1:
                matches = matches[matches[:, 2].argsort()[::-1]]
                matches = matches[np.unique(matches[:, 1], return_index=True)[1]]
                matches = matches[matches[:, 2].argsort()[::-1]]
                matches = matches[np.unique(matches[:, 0], return_index=True)[1]]
        else:
            matches = np.zeros((0, 3))
        n = matches.shape[0] > 0
        m0, m1, _ = matches.transpose().astype(int)
        for i, gc in enumerate(gt_classes):
            j = m0 == i
            if n and sum(j) == 1:
                self.matrix_mask[detection_classes[m1[j]], gc] += 1  # correct
            else:
                self.matrix_mask[self.nc, gc] += 1  # true background
        if n:
            for i, dc in enumerate(detection_classes):
                if not any(m1 == i):
                    self.matrix_mask[dc, self.nc] += 1  # predicted background
    def eval_mask_combine(self,mask_detection, gt_cls,mask_gt,id_name=None):
        #no label --> predict (thua) , label -- > no predict(thieu)
        mask_detection = mask_detection[mask_detection[:, -2] > self.conf]
        if mask_detection.shape[0] == 0: mask_detection = None
        self.total_object += mask_gt.shape[0]
        if gt_cls.shape[0] == 0:  # Check if labels is empty
            if mask_detection is not None:
                self.total_mask_predict += mask_detection.shape[0]
                mask_detection = mask_detection[mask_detection[:, -2] > self.conf]
                detection_classes = mask_detection[:, -1].int()
                for dc in detection_classes:
                    self.sum_surplus +=1
            return
        if mask_detection is None:
            gt_classes = gt_cls.int()
            for gc in gt_classes:
                self.sum_shortage+=1
            return
        # wrong_detect,miss_detect = bgr_check(gt_cls,detections)
        iou = box_iou(mask_gt, mask_detection[:, :-2])
        correct_count = self.count_correct(iou,mask_gt,mask_detection[:, :-2])
        # correct_count = torch.count_nonzero(torch.sum(iou, dim=1) > self.iou_thres).item()
        iou_each_mask = iou_smask(mask_gt, mask_detection[:, :-2])
        sum_mask_pr = iou_each_mask.shape[1]
        mask_acp = torch.count_nonzero(torch.sum(iou_each_mask, dim=0) >= self.mask_accept).item()
        self.total_mask_predict += sum_mask_pr
        self.total_mask_accept += mask_acp

        shortage_count = torch.count_nonzero(torch.sum(iou, dim=1) == 0).item()
        shortage_iou_count = torch.count_nonzero((torch.sum(iou, dim=1) > 0) & (torch.sum(iou, dim=1) < self.iou_thres)).item()
        surplus_count = torch.count_nonzero(torch.sum(iou, dim=0) == 0).item()
        self.sum_correct += correct_count

        self.sum_shortage_iou += shortage_iou_count
        self.sum_shortage += shortage_count
        self.sum_surplus += surplus_count
        if id_name is not None:
            if surplus_count != 0:
                self.name_surplus.append(id_name)
            elif shortage_iou_count != 0 or shortage_count != 0:
                self.name_shortage.append(id_name)

    def process_batch(self, detections, gt_bboxes, gt_cls):
        """
        Update confusion matrix for object detection task.

        Args:
            detections (Array[N, 6] | Array[N, 7]): Detected bounding boxes and their associated information.
                                      Each row should contain (x1, y1, x2, y2, conf, class)
                                      or with an additional element `angle` when it's obb.
            gt_bboxes (Array[M, 4]| Array[N, 5]): Ground truth bounding boxes with xyxy/xyxyr format.
            gt_cls (Array[M]): The class labels.
        """
        if gt_cls.shape[0] == 0:  # Check if labels is empty
            if detections is not None:
                detections = detections[detections[:, 4] > self.conf]
                detection_classes = detections[:, 5].int()
                for dc in detection_classes:
                    self.matrix[dc, self.nc] += 1  # false positives
            return
        if detections is None:
            gt_classes = gt_cls.int()
            for gc in gt_classes:
                self.matrix[self.nc, gc] += 1  # background FN
            return

        detections = detections[detections[:, 4] > self.conf]
        gt_classes = gt_cls.int()
        detection_classes = detections[:, 5].int()
        is_obb = detections.shape[1] == 7 and gt_bboxes.shape[1] == 5  # with additional `angle` dimension
        iou = (
            batch_probiou(gt_bboxes, torch.cat([detections[:, :4], detections[:, -1:]], dim=-1))
            if is_obb
            else box_iou(gt_bboxes, detections[:, :4])
        )
        x = torch.where(iou > self.iou_thres)
        if x[0].shape[0]:
            matches = torch.cat((torch.stack(x, 1), iou[x[0], x[1]][:, None]), 1).cpu().numpy()
            if x[0].shape[0] > 1:
                matches = matches[matches[:, 2].argsort()[::-1]]
                matches = matches[np.unique(matches[:, 1], return_index=True)[1]]
                matches = matches[matches[:, 2].argsort()[::-1]]
                matches = matches[np.unique(matches[:, 0], return_index=True)[1]]
        else:
            matches = np.zeros((0, 3))

        n = matches.shape[0] > 0
        m0, m1, _ = matches.transpose().astype(int)
        for i, gc in enumerate(gt_classes):
            j = m0 == i
            if n and sum(j) == 1:
                self.matrix[detection_classes[m1[j]], gc] += 1  # correct
            else:
                self.matrix[self.nc, gc] += 1  # true background

        if n:
            for i, dc in enumerate(detection_classes):
                if not any(m1 == i):
                    self.matrix[dc, self.nc] += 1  # predicted background
    def count_correct(self,iou,mask_gt,mask_detection):
        if iou.shape[0] != 0:
            max_values, _ = torch.max(iou, dim=1)
            count_true = (max_values > self.iou_thres).sum().item()
        if torch.max((iou != 0).sum(dim=1)) > 1:
            indices_list = [(iou[row] != 0).nonzero(as_tuple=True)[0] for row in range(iou.size(0))]
            for i,gt in enumerate(indices_list):
                if gt.shape[0] != 1 and 0 < torch.max(iou[i]) < self.iou_thres:
                    mdt = mask_detection[gt]
                    mask_merge = torch.sum(mdt,dim=0)
                    mask_merge[mask_merge!=0] = 1
                    gt_multi_mask = mask_gt[i].unsqueeze(0)
                    multi_mask = mask_merge.unsqueeze(0)
                    iou_mask = box_iou(gt_multi_mask,multi_mask)
                    if iou_mask > self.iou_thres: count_true+=1
        return count_true
    def matrix(self):
        """Returns the confusion matrix."""
        return self.matrix
    def matrix_mask(self):
        """Returns the confusion matrix."""
        return self.matrix_mask
    def tp_fp(self):
        """Returns true positives and false positives."""
        tp = self.matrix.diagonal()  # true positives
        fp = self.matrix.sum(1) - tp  # false positives
        # fn = self.matrix.sum(0) - tp  # false negatives (missed detections)
        return (tp[:-1], fp[:-1],) if self.task == "detect" else (tp, fp)  # remove background class if task=detect

    def plot(self, normalize=True, save_dir="", names=(), on_plot=None):
        """
        Plot the confusion matrix using seaborn and save it to a file.

        Args:
            normalize (bool): Whether to normalize the confusion matrix.
            save_dir (str): Directory where the plot will be saved.
            names (tuple): Names of classes, used as labels on the plot.
            on_plot (func): An optional callback to pass plots path and data when they are rendered.
        """
        import seaborn  # scope for faster 'import ultralytics'

        array = self.matrix / ((self.matrix.sum(0).reshape(1, -1) + 1e-9) if normalize else 1)  # normalize columns
        array[array < 0.005] = np.nan  # don't annotate (would appear as 0.00)

        fig, ax = plt.subplots(1, 1, figsize=(12, 9), tight_layout=True)
        nc, nn = self.nc, len(names)  # number of classes, names
        seaborn.set_theme(font_scale=1.0 if nc < 50 else 0.8)  # for label size
        labels = (0 < nn < 99) and (nn == nc)  # apply names to ticklabels
        ticklabels = (list(names) + ["background"]) if labels else "auto"
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")  # suppress empty matrix RuntimeWarning: All-NaN slice encountered
            seaborn.heatmap(
                array,
                ax=ax,
                annot=nc < 30,
                annot_kws={"size": 8},
                cmap="Blues",
                fmt=".2f" if normalize else ".0f",
                square=True,
                vmin=0.0,
                xticklabels=ticklabels,
                yticklabels=ticklabels,
            ).set_facecolor((1, 1, 1))
        title = "Confusion Matrix" + " Normalized" * normalize
        ax.set_xlabel("True")
        ax.set_ylabel("Predicted")
        ax.set_title(title)
        plot_fname = Path(save_dir) / f'{title.lower().replace(" ", "_")}.png'
        fig.savefig(plot_fname, dpi=250)
        plt.close(fig)
        if on_plot:
            on_plot(plot_fname)
    def print(self):
        for i in range(self.nc + 1):
            print(' '.join(map(str, self.matrix[i])))
    def plot_cfs(self,matrix_type="detect",normalize=True,  names=(), on_plot=None):
        if matrix_type == 'mask':
            matrix = self.matrix_mask
        else:
            matrix = self.matrix
        import seaborn  # scope for faster 'import ultralytics'
        array = matrix/ ((matrix.sum(0).reshape(1, -1) + 1e-9) if normalize else 1)  # normalize columns
        array[array < 0.005] = np.nan  # don't annotate (would appear as 0.00)
        fig, ax = plt.subplots(1, 1, figsize=(12, 9), tight_layout=True)
        nc, nn = self.nc, len(names)  # number of classes, names
        seaborn.set_theme(font_scale=1.0 if nc < 50 else 0.8)  # for label size
        labels = (0 < nn < 99) and (nn == nc)  # apply names to ticklabels
        ticklabels = (list(names) + ["background"]) if labels else "auto"
        with warnings.catch_warnings():
            warnings.simplefilter("ignore")  # suppress empty matrix RuntimeWarning: All-NaN slice encountered
            seaborn.heatmap(
                array,
                ax=ax,
                annot=nc < 30,
                annot_kws={"size": 8},
                cmap="Blues",
                fmt=".2f" if normalize else ".0f",
                square=True,
                vmin=0.0,
                xticklabels=ticklabels,
                yticklabels=ticklabels,
            ).set_facecolor((1, 1, 1))
        title = "Confusion Matrix" + " Normalized" * normalize
        ax.set_xlabel("True")
        ax.set_ylabel("Predicted")
        ax.set_title(title)
        plt.show()
def get_gt(label_path, results):
    h, w = results.orig_shape
    gt_bboxes = []
    gt_cls = []
    try:
        with open(label_path, 'r') as label:
            objects = label.readlines()
        for line in objects:
            class_id, cx, cy, bw, bh = map(float, line.strip().split())
            xmin = round((cx - bw / 2) * w)
            xmax = round((cx + bw / 2) * w)
            ymin = round((cy - bh / 2) * h)
            ymax = round((cy + bh / 2) * h)
            box = [xmin, ymin, xmax, ymax]
            gt_bboxes.append(box)
            gt_cls.append(int(class_id))
    except Exception as e:
        print(f"Error reading {label_path}: {e}")

    gt_bboxes = torch.tensor(gt_bboxes, dtype=torch.float)
    gt_cls = torch.tensor(gt_cls, dtype=torch.int)
    detections = torch.cat((
        results.boxes.xyxy,
        results.boxes.conf.unsqueeze(1),
        results.boxes.cls.unsqueeze(1)), dim=1)

    return gt_bboxes, gt_cls, detections

def resize_mask(mask, image):
    h, w = mask.shape
    H, W, _ = image.shape
    if (w, h) == (W, H):
        return mask
    gain = min(w / W, h / H)  # gain  = old / new
    pad = (w - W * gain) / 2, (h - H * gain) / 2  # wh padding
    top, left = int(pad[1]), int(pad[0])  # y, x
    bottom, right = int(h - pad[1]), int(w - pad[0])
    mask = mask[top:bottom, left:right]
    mask = mask[None, None, :, :]
    mask = F.interpolate(mask, size=(H, W))
    mask = mask.squeeze()
    return mask

def polygon2mask(imgsz, polygons, color=1, downsample_ratio=1):
    mask = np.zeros(imgsz, dtype=np.uint8)
    polygons = np.asarray(polygons, dtype=np.int32)
    polygons = polygons.reshape((polygons.shape[0], -1, 2))
    cv2.fillPoly(mask, [polygons], color=color)
    nh, nw = (imgsz[0] // downsample_ratio, imgsz[1] // downsample_ratio)
    return cv2.resize(mask, (nw, nh))
def polygons2masks(imgsz, polygons, color=255, downsample_ratio=1):
    return np.array([polygon2mask(imgsz, x.reshape(-1, 2), color, downsample_ratio) for x in polygons])
def precision_score_(groundtruth_mask, pred_mask):
    intersect = torch.sum(pred_mask * groundtruth_mask)
    total_pixel_pred = torch.sum(pred_mask)
    precision = intersect.float() / total_pixel_pred.float()
    return round(precision.item(), 3)

def recall_score_(groundtruth_mask, pred_mask):
    intersect = torch.sum(pred_mask * groundtruth_mask)
    total_pixel_truth = torch.sum(groundtruth_mask)
    recall = intersect.float() / total_pixel_truth.float()
    return round(recall.item(), 3)

def iou(groundtruth_mask, pred_mask):
    intersect = torch.sum(pred_mask*groundtruth_mask)
    union = torch.sum(pred_mask) + torch.sum(groundtruth_mask) - intersect
    iou = torch.mean(intersect/union)
    return round(iou.item(), 3)
def metrics(groundtruth_mask, pred_mask):
    precision = precision_score_(groundtruth_mask, pred_mask)
    recall = recall_score_(groundtruth_mask, pred_mask)
    # accuracy = accuracy(groundtruth_mask, pred_mask)
    iou_ = iou(groundtruth_mask, pred_mask)
    return precision,recall,iou_
def eval_img(image_path,model=model):
    res = model.predict(image_path,verbose=False)[0]
    if res.masks is not None:
        mask = res.masks.data
        mask = torch.any(res.masks.data, dim=0).to(torch.uint8)
        pred_masks = resize_mask(mask, res.orig_img)
    else: pred_masks = torch.zeros(res.orig_shape)
    height, width = res.orig_shape
    try:
        polygons = []
        with open(image_path.replace('.png','.txt').replace('images','labels'), "r") as file_txt:
            datas = file_txt.readlines()
            for line in datas:
                polygon = []
                points = [float(coord) for coord in line.strip().split(" ")[1:]]
                x_coords, y_coords = np.array(points[::2])*width, np.array(points[1::2])*height
                polygon = np.column_stack((x_coords, y_coords)).astype(int).flatten()
                polygons.append(polygon)
        gt_masks = polygons2masks(res.orig_shape, polygons, color=255, downsample_ratio=1)
        gt_masks =  torch.tensor(gt_masks)
        gt_masks = torch.any(gt_masks, dim=0).to(torch.uint8)
        gt_masks[gt_masks!=0] = 1
    except:
        gt_masks = torch.zeros(res.orig_shape)
    gt_masks = gt_masks.to(device)
    pred_masks = pred_masks.to(device)
    if torch.all(gt_masks==0) and torch.all(pred_masks==0):
        return 1,1,1
    precision,recall,iou_ = metrics(gt_masks, pred_masks)
    return precision,recall,iou_

def get_mask_and_img(image_path,model=model):
    res = model.predict(image_path,verbose =False)[0]
    if res.masks is not None:
        mask = res.masks.data
        mask = torch.any(res.masks.data, dim=0).to(torch.uint8)
        pred_masks = resize_mask(mask, res.orig_img)
    else: pred_masks = torch.zeros(res.orig_shape)
    height, width = res.orig_shape
    try:
        polygons = []
        with open(image_path.replace('.png','.txt').replace('images','labels'), "r") as file_txt:
            datas = file_txt.readlines()
            for line in datas:
                polygon = []
                points = [float(coord) for coord in line.strip().split(" ")[1:]]
                x_coords, y_coords = np.array(points[::2])*width, np.array(points[1::2])*height
                polygon = np.column_stack((x_coords, y_coords)).astype(int).flatten()
                polygons.append(polygon)
        gt_masks = polygons2masks(res.orig_shape, polygons, color=255, downsample_ratio=1)
        gt_masks =  torch.tensor(gt_masks)
        gt_masks = torch.any(gt_masks, dim=0).to(torch.uint8)
        gt_masks[gt_masks!=0] = 1
    except:
        gt_masks = torch.zeros(res.orig_shape)
    gt_masks = gt_masks.to(device)
    pred_masks = pred_masks.to(device)
    if torch.all(gt_masks==0) and torch.all(pred_masks==0):
        return pred_masks,gt_masks,res.orig_img,1,1,1
    precision,recall,iou_ = metrics(gt_masks, pred_masks)
    return pred_masks,gt_masks,res.orig_img,precision,recall,iou_
def create_mask_from_polygon(polygon_points, image_shape):
    mask = np.zeros(image_shape, dtype=np.uint8)
    points = np.array([polygon_points], dtype=np.int32)
    cv2.fillPoly(mask, points, color=1)
    return mask
def polygons_to_tensor_masks(polygons, image_shape):
    # Pre-allocate a NumPy array for all masks
    num_polygons = len(polygons)
    masks = np.zeros((num_polygons, np.prod(image_shape)), dtype=np.uint8)
    for i, polygon in enumerate(polygons):
        mask = create_mask_from_polygon(polygon, image_shape)
        masks[i] = mask.flatten()
    masks_tensor = torch.tensor(masks, dtype=torch.float32)
    return masks_tensor
def get_mask_gt(label_path,results):
    if results.masks is None:
        polygons_detection = []
    else :
        polygons_detection = results.masks.xy
    try:
        polygons_gt = txt2point(label_path, results.orig_shape)
    except:
        polygons_gt = []
    mask_detection = polygons_to_tensor_masks(polygons_detection, results.orig_shape).to(device)
    mask_gt = polygons_to_tensor_masks(polygons_gt, results.orig_shape).to(device)
    mask_detection = torch.cat((mask_detection, results.boxes.conf .unsqueeze(1), results.boxes.cls .unsqueeze(1)), dim=1)
    return mask_detection, mask_gt

def multi_plot_cfs(cfs_matrics, matrix_type="detect", normalize=True, names=(), on_plot=None):
    num_matrices = len(cfs_matrics)
    fig, axs = plt.subplots(1, num_matrices, figsize=(14 * num_matrices, 10), tight_layout=True)
    if num_matrices == 1:
        axs = [axs]
    for idx, key in enumerate(cfs_matrics):
        if matrix_type == "detect":
            matrix = cfs_matrics[key].matrix
        else:
            matrix = cfs_matrics[key].matrix_mask
        array = matrix / ((matrix.sum(0).reshape(1, -1) + 1e-9) if normalize else 1)  # normalize columns
        array[array < 0.005] = np.nan  # don't annotate (would appear as 0.00)
        ax = axs[idx]
        nc, nn = len(matrix), len(names)+1  # number of classes, names
        sns.set_theme(font_scale=1.0 if nc < 50 else 0.8)  # for label size
        labels = (0 < nn < 99) and (nn == nc)  # apply names to ticklabels
        ticklabels = names if labels else "auto"
        sns.heatmap(
            array,
            ax=ax,
            annot=nc < 30,
            annot_kws={"size": 8},
            cmap="Blues",
            fmt=".2f" if normalize else ".0f",
            square=True,
            vmin=0.0,
            xticklabels=ticklabels,
            yticklabels=ticklabels,
        ).set_facecolor((1, 1, 1))
        title = f"Confusion Matrix {key.split('_')[-1]} of {matrix_type}" + " Normalized" * normalize
        ax.set_xlabel("True")
        ax.set_ylabel("Predicted")
        ax.set_title(title)
    plt.show()

def get_error_mask(mask_detection,mask_gt,results,mask_size=(1080, 1920)):
    mask_detection = mask_detection[mask_detection[:, -2] > 0.25]
    iou = box_iou(mask_gt, mask_detection[:, :-2])
    precison_error_mask = torch.sum(iou,dim=0)
    recall_error_mask = torch.sum(iou,dim=1)
    precison_error_mask_indeces = torch.nonzero(precison_error_mask == 0).squeeze() #predict
    recall_error_mask_indeces = torch.nonzero(recall_error_mask == 0).squeeze() #gt
    error_masks_recall = mask_gt[recall_error_mask_indeces].view(-1, results.orig_shape[0],results.orig_shape[1])
    error_masks_precision = mask_detection[precison_error_mask_indeces, :-2].view(-1, results.orig_shape[0],results.orig_shape[1])
    return error_masks_recall,error_masks_precision

def get_image_wrong():
    # missing_multi_mask_nu += torch.sum(torch.sum(iou != 0, dim=1) > 1).item()
    # x = torch.where(torch.logical_and(iou > 0, iou < 0.45))
    # missing_iou_nu+= x[0].numel()
    # if x[0].numel() != 0:
    #     name_wrong.append(img)
    # if torch.sum(torch.sum(iou != 0, dim=1) > 1).item() != 0:
    #     name_miss.append(img)
    return
def bgr_check(gt_cls,detections,conf=0.25):
    wrong_detect = 0
    miss_detect = 0
    if gt_cls.shape[0] == 0:  # Check if labels is empty
        if detections is not None:
            detections = detections[detections[:, 4] > conf]
            detection_classes = detections[:, 5].int()
            for dc in detection_classes:
                wrong_detect +=1
    if detections is None:
        gt_classes = gt_cls.int()
        for gc in gt_classes:
            miss_detect+=1
    return wrong_detect,miss_detect

def get_scores_metrics(cfs_matrix,task="detect"):
    if task == "detect":
        matrix = cfs_matrix.matrix
    else:
        matrix = cfs_matrix.matrix_mask
    tp = cfs_matrix.sum_correct
    fp = cfs_matrix.sum_surplus
    fn = cfs_matrix.sum_shortage + cfs_matrix.sum_shortage_iou
    tn = 0
    accuracy = (tp + tn) /  cfs_matrix.total_object
    precision = tp / (tp+fp)
    recall = tp / (tp+fn)
    f1 = precision*recall*2/(precision+recall)
    total_actual_objects = cfs_matrix.total_object
    # total_mispredicted_objects = np.sum(matrix[-1,:]) +  np.sum(matrix[:,-1])
    data = {
        'Task': [task],
        'IoU': [float(cfs_matrix.iou_thres)],
        'Conf': [float(cfs_matrix.conf)],
        'Accuracy': [float(accuracy)],
        'Precision': [float(precision)],
        'Recall': [float(recall)],
        'F1-score': [float(f1)],
        'Total Objects': [int(cfs_matrix.total_object)],
        'Sum correct': [int(cfs_matrix.sum_correct)],
        'Sum shortage iou': [int(cfs_matrix.sum_shortage_iou)],
        'Sum shortage': [int(cfs_matrix.sum_shortage)],
        'Sum surplus': [int(cfs_matrix.sum_surplus)],
        # 'Percent bet covered':[cfs_matrix.sum_correct/cfs_matrix.total_object],
    }
    return data
def get_box_and_img(image_path, model=model):
    res = model.predict(image_path, verbose=False)[0]

    # 1. Predict boxes
    if res.boxes is not None and res.boxes.shape[0] > 0:
        pred_boxes = res.boxes.data[:, :4]  # x1, y1, x2, y2
    else:
        pred_boxes = torch.empty((0, 4))

    # 2. Ground truth boxes
    height, width = res.orig_shape
    try:
        gt_boxes = []
        with open(image_path.replace('.png', '.txt').replace('images', 'labels'), "r") as file_txt:
            for line in file_txt.readlines():
                points = [float(p) for p in line.strip().split(" ")[1:]]
                x_coords = np.array(points[::2]) * width
                y_coords = np.array(points[1::2]) * height
                x1, y1 = np.min(x_coords), np.min(y_coords)
                x2, y2 = np.max(x_coords), np.max(y_coords)
                gt_boxes.append([x1, y1, x2, y2])
        gt_boxes = torch.tensor(gt_boxes, dtype=torch.float32)
    except:
        gt_boxes = torch.empty((0, 4))

    # 3. Move to device
    gt_boxes = gt_boxes.to(device)
    pred_boxes = pred_boxes.to(device)

    # 4. Check empty
    if pred_boxes.shape[0] == 0 and gt_boxes.shape[0] == 0:
        return pred_boxes, gt_boxes, res.orig_img, 1, 1, 1

    # 5. Metric with IoU
    iou_matrix = box_iou(gt_boxes, pred_boxes)
    matched_iou = torch.max(iou_matrix, dim=1).values
    matched_pred = torch.max(iou_matrix, dim=0).values

    tp = (matched_iou > 0.5).sum().item()
    fp = (pred_boxes.shape[0] - tp)
    fn = (gt_boxes.shape[0] - tp)

    precision = tp / (tp + fp + 1e-6)
    recall = tp / (tp + fn + 1e-6)
    mean_iou = matched_iou[matched_iou > 0.5].mean().item() if (matched_iou > 0.5).sum() > 0 else 0

    return pred_boxes, gt_boxes, res.orig_img, precision, recall, mean_iou
def transfer_tensor(path_img,size=(640, 640)):
    resizer = torchvision.transforms.Resize(size=size)
    frames = cv2.imread(path_img)
    data = torch.from_numpy(frames).to(device)
    data = [data]
    batch_tensor = torch.stack(data)
    batch_tensor = batch_tensor.permute(0, 3, 1, 2)
    batch_tensor = resizer(batch_tensor)
    batch_tensor = batch_tensor.float() / 255.0
    return batch_tensor
```

```python
import torch
import numpy as np

def box_iou(boxes1, boxes2):
    """
    Compute IoU between two sets of boxes.
    boxes1: Tensor of shape (N, 4) - Ground truth boxes (x1, y1, x2, y2)
    boxes2: Tensor of shape (M, 4) - Predicted boxes (x1, y1, x2, y2)
    Returns: IoU matrix of shape (N, M)
    """
    area1 = (boxes1[:, 2] - boxes1[:, 0]) * (boxes1[:, 3] - boxes1[:, 1])
    area2 = (boxes2[:, 2] - boxes2[:, 0]) * (boxes2[:, 3] - boxes2[:, 1])

    lt = torch.max(boxes1[:, None, :2], boxes2[:, :2])  # [N, M, 2]
    rb = torch.min(boxes1[:, None, 2:], boxes2[:, 2:])  # [N, M, 2]

    wh = (rb - lt).clamp(min=0)  # [N, M, 2]
    inter = wh[:, :, 0] * wh[:, :, 1]  # [N, M]

    union = area1[:, None] + area2 - inter
    iou = inter / (union + 1e-6)
    return iou

def get_box_and_img(image_path, model, device, target_class=2):
    res = model.predict(image_path, verbose=False)[0]

    # 1. Predict boxes for class 2
    if res.boxes is not None and res.boxes.shape[0] > 0:
        pred_data = res.boxes.data  # [x1, y1, x2, y2, conf, class_id]
        pred_classes = pred_data[:, 5].int()  # Class IDs
        pred_mask = pred_classes == target_class  # Mask for class 2
        pred_boxes = pred_data[pred_mask, :4]  # Boxes for class 2: x1, y1, x2, y2
        pred_class_indices = torch.where(pred_mask)[0]  # Indices for class 2 in original data
    else:
        pred_boxes = torch.empty((0, 4))
        pred_class_indices = torch.tensor([])

    # 2. Ground truth boxes and classes for class 2
    height, width = res.orig_shape
    try:
        gt_boxes = []
        gt_classes = []
        label_path = image_path.replace('.jpg', '.txt').replace('images', 'labels')
        with open(label_path, "r") as file_txt:
            for line in file_txt.readlines():
                class_id, x_center, y_center, w, h = [float(p) for p in line.strip().split()]
                if int(class_id) == target_class:  # Chỉ lấy class 2
                    x_center, y_center, w, h = x_center * width, y_center * height, w * width, h * height
                    x1 = x_center - w / 2
                    y1 = y_center - h / 2
                    x2 = x_center + w / 2
                    y2 = y_center + h / 2
                    gt_boxes.append([x1, y1, x2, y2])
                    gt_classes.append(int(class_id))
        gt_boxes = torch.tensor(gt_boxes, dtype=torch.float32)
        gt_classes = torch.tensor(gt_classes, dtype=torch.int32)
    except FileNotFoundError:
        print(f"Label file not found: {label_path}")
        gt_boxes = torch.empty((0, 4))
        gt_classes = torch.empty((0,), dtype=torch.int32)
    except Exception as e:
        print(f"Error reading label file: {e}")
        gt_boxes = torch.empty((0, 4))
        gt_classes = torch.empty((0,), dtype=torch.int32)

    # 3. Move to device
    gt_boxes = gt_boxes.to(device)
    pred_boxes = pred_boxes.to(device)
    gt_classes = gt_classes.to(device)

    # 4. Check if either is empty
    if pred_boxes.shape[0] == 0 and gt_boxes.shape[0] == 0:
        return pred_boxes, gt_boxes, gt_classes, res.orig_img, 1.0, 1.0, 1.0, [], []
    elif pred_boxes.shape[0] == 0 or gt_boxes.shape[0] == 0:
        fn_indices = list(range(gt_boxes.shape[0])) if pred_boxes.shape[0] == 0 else []  # Bắt thiếu
        fp_indices = list(range(pred_boxes.shape[0])) if gt_boxes.shape[0] == 0 else []  # Bắt thừa
        return pred_boxes, gt_boxes, gt_classes, res.orig_img, 0.0, 0.0, 0.0, fn_indices, fp_indices

    # 5. Compute IoU for class 2
    iou_matrix = box_iou(gt_boxes, pred_boxes)
    max_iou, match_indices = torch.max(iou_matrix, dim=1)
    matched = max_iou > 0.5  # Ngưỡng IoU

    tp = matched.sum().item()
    fn_indices = torch.where(~matched)[0].tolist()  # Ground truth không được dự đoán (FN)
    fp_indices = [i for i in range(pred_boxes.shape[0]) if i not in match_indices[matched].tolist()]  # Dự đoán không khớp (FP)

    precision = tp / (tp + len(fp_indices) + 1e-6)
    recall = tp / (tp + len(fn_indices) + 1e-6)
    mean_iou = max_iou[matched].mean().item() if tp > 0 else 0.0

    return pred_boxes, gt_boxes, gt_classes, res.orig_img, precision, recall, mean_iou, fn_indices, fp_indices

from PIL import ImageDraw, Image

def draw_boxes_on_image(image, boxes, classes=None, fn_indices=None, fp_indices=None, target_class=2, default_color=(0, 255, 0), error_color=(255, 255, 0), width=3, is_pred=False):
    """Draw bounding boxes with error highlighting for class 2"""
    image = Image.fromarray(image.astype('uint8')).convert("RGB")
    draw = ImageDraw.Draw(image)

    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = box.tolist()
        if is_pred:
            # For predicted boxes
            color = (0, 0, 255) if i in fp_indices else (255, 0, 0)  # Blue for FP, Red for others
        else:
            # For ground truth boxes
            color = error_color if i in fn_indices and classes[i].item() == target_class else default_color  # Yellow for FN, Green for others
        draw.rectangle([x1, y1, x2, y2], outline=color, width=width)

    return np.array(image)

import os
from tqdm import tqdm
import matplotlib.pyplot as plt
print(recall_error)
threshout = 0.25
IMAGES_PATH = "/content/drive/MyDrive/AIP491/Training/yolo_helmet/data_mixup/images/train/"
data_set = os.listdir(IMAGES_PATH)[651:]
# Assume cfs_matrics is defined; otherwise, remove related conditions
try:
    first_conf_matrix = list(cfs_matrics.values())[0]
except NameError:
    first_conf_matrix = None

for img in tqdm(data_set):
    if img.startswith("aug_mixup"):
      continue
    # Get boxes, classes, metrics, and error indices
    pred_boxes, gt_boxes, gt_classes, image, precision, recall, iou_, fn_indices, fp_indices = get_box_and_img(
        f'{IMAGES_PATH}{img}', model=model, device="cuda", target_class=2
    )

    # Convert image from torch to numpy if needed
    if isinstance(image, torch.Tensor):
        image = image.permute(1, 2, 0).cpu().numpy()

    # Draw boxes on image
    img_with_gt = draw_boxes_on_image(
        image,
        gt_boxes,
        classes=gt_classes,
        fn_indices=fn_indices,
        default_color=(0, 255, 0),  # Green for non-error GT
        error_color=(255, 255, 0),  # Yellow for FN of class 2
        is_pred=False
    )
    img_with_pred = draw_boxes_on_image(
        image,
        pred_boxes,
        fp_indices=fp_indices,
        default_color=(255, 0, 0),  # Red for non-error predictions
        error_color=(0, 0, 255),   # Blue for FP of class 2
        is_pred=True
    )

    # Plot
    plt.figure(figsize=(10, 5))

    # Ensure RGB before showing (in case image is BGR)
    img_with_gt = img_with_gt[:, :, ::-1]
    img_with_pred = img_with_pred[:, :, ::-1]

    plt.subplot(1, 2, 1)
    plt.imshow(img_with_gt)
    plt.title('Ground Truth Boxes')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(img_with_pred)
    plt.title('Predicted Boxes')
    plt.axis('off')

    # Save images based on conditions
    if recall < threshout or fn_indices:
        print(f"Low Recall or FN for class 2 → Save recall error: {img}")
        plt.savefig(f'{recall_error}{img}', dpi=300)
    if precision < threshout or fp_indices:
        print(f"Low Precision or FP for class 2 → Save precision error: {img}")
        plt.savefig(f'{precision_error}{img}', dpi=300)

    # if first_conf_matrix and img in first_conf_matrix.name_shortage:
    #     plt.savefig(f'{recall_error}{img}', dpi=300)

    # if first_conf_matrix and img in first_conf_matrix.name_surplus:
    #     plt.savefig(f'{precision_error}{img}', dpi=300)

    plt.close()
```
